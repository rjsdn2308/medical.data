# HTTP RequestÎ•º ÌôúÏö©Ìïú Îç∞Ïù¥ÌÑ∞ ÏàòÏßë

```python
import requests, sys

# ÏöîÏ≤≠ url ÏÑ§Ï†ï
server_url = "https://rest.ensembl.org"
ext_url = "/overlap/id/ENSG00000157764?feature=gene"

# HTTP GET method Ìò∏Ï∂ú
r = requests.get(server_url + ext_url, headers={ "Content-Type" : "application/json"})

# ÏòàÏô∏ Ï≤òÎ¶¨
if not r.ok:
  r.raise_for_status()
  sys.exit()

# Response ÌôïÏù∏
decoded = r.json()
print(repr(decoded))
```

```python
# Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑ ÌòïÌÉúÎ°ú Ï†ÄÏû• Î∞è Ï∂úÎ†•
import pandas as pd
df = pd.DataFrame(decoded)
df
```
